```python
df_X=df.iloc[:,:4]
df_Y=df.iloc[:,4:5]
```

```python
df_X_train,df_X_test,df_Y_train,df_Y_test=train_test_split(df_X,df_Y,test_size=0.2)
```

```python
lr=LinearRegression()
```

```python
lr.fit(df_X_train,df_Y_train)
```

```python
print(lr.intercept_,lr.coef_)
```

For a given amount of Radio and Newspaper ad spending, an increase of $1000 in TV ad spending is associated with an increase in Sales of 45.765 widgets.

```python
lr.score(df_X_train,df_Y_train)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
accuracies = cross_val_score(estimator=lr, X=df_X_train, y=df_Y_train, cv=10)
model_accuracy = accuracies.mean()
model_standard_deviation = accuracies.std()
```
